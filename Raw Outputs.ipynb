```python
from youtube_api.google.google_api import ApiKeyManager, get_keys, ResourceManager
```

```python
api_key_manager = ApiKeyManager()
resource_manager = ResourceManager(api_key_manager)
```

## Response Data Structures for Each Endpoint

```python
resource = resource_manager.get_resource()
```

```python
# for testing
channel_id = 'UCgrNz-aDmcr2uuto8_DL2jg'  # CGTN
video_id = 'CIzSm1O2oCo'  # Some CGTN video
```

### Get Channel

```python
fn = resource.channels().list
kwargs = {
    'id': channel_id,
    'part': 'contentDetails,snippet,statistics,topicDetails',
}
request = fn(**kwargs)
response = request.execute()
```

### Get Channel Uploads Stream (New Function)

```python
request = resource.channels().list(
    part='contentDetails',
    id=channel_id)
response = request.execute()
stream = response['items'][0]['contentDetails']['relatedPlaylists'][
    'uploads']
stream
```

### Get Channel Videos

```python
fn = resource.playlistItems().list
kwargs = {
    'part': 'snippet',
    'playlistId': stream,
}
response = fn(**kwargs).execute()
response
```

### Get Video Comments

```python
fn = resource.commentThreads().list
kwargs = dict(
    part='snippet,replies',
    maxResults=100,
    videoId=video_id,
)
response = fn(**kwargs).execute()
response
```

### Get Video

```python
fn = resource.videos().list
kwargs = dict(
    part='snippet,contentDetails,statistics',
    id=video_id
)
response = fn(**kwargs).execute()
response
```

### Search

```python
fn = resource.search().list
kwargs = dict(
    part='snippet',
    q='Pomegranates',
    channelId=channel_id,
    maxResults=50,
    order='rating',
    type='video',
    publishedAfter='2023-09-01T00:00:00Z',
    publishedBefore='2023-10-15T00:00:00Z'
)
response = fn(**kwargs).execute()
response
```
